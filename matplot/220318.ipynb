# 표준 정규분포 그래프

```python
x = np.linspace(-5,5,101)
y = (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)
plt.plot(x,y)
plt.xlabel("x")  # 표준 정규분포 함수이다.
plt.ylabel("y")
plt.show()
```

```python
import scipy.stats as stats
```

```python
y_ = stats.norm(0,1).pdf(x) # norm(평균, 표준편차)
```

```python
def norm(_a,_b,_x):
    return stats.norm(_a,_b).pdf(_x)
x= np.linspace(-20,20,101)
for i in range(1,11,1):
    _y = norm(0,i,x)
    plt.plot(x,_y)
plt.show()
   
```

# 파이차트

```python
x = np.random.randint(10,100,5)
labels = ["A", "B", "C", "D", "E"]
```

```python
explodes =[0,0.10,0.30,0.20,0.10] # 차트 나눠서 표현
plt.pie(x, autopct='%.1f%%', labels=labels,  # 퍼센트 소수점 첫째자리까지 표시
        counterclock = True, startangle=180, explode=explodes)  # true 가 반시계 방향 , false가 시계방향
plt.show() 
```

```python
x = np.linspace(0,10,10)
y = np.random.rand(10)
```

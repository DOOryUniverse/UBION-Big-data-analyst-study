```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['axes.grid'] = True
```

```python
import pandas as pd
import pandas_datareader as pdr
```

```python
df_gold = pdr.DataReader('GOLDAMGBD228NLBM', 'fred', start='2000-01-01')
print(type(df_gold))
print('row count:', len(df_gold))
```

```python
# 일반 숫자로 그래프 그려보기
ar_process = []
y = 140
coff = 0.3
const = 100
for i in range(300):
 error = np.random.randn()
 y = const + y * coff + error
 ar_process.append(y)
print(const / (1 - coff))
plt.plot(ar_process)
```

```python
# 함수 만들기
def plot_rolling(data, interval):
    rolmean = data.rolling(interval, min_periods=1).mean()
    rolstd = data.rolling(interval, min_periods=1).std()
    print(rolmean)
    #Plot rolling statistics:
    plt.figure(figsize=(10, 6))
    plt.xlabel('Date')
    orig = plt.plot(data, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.show()
plot_rolling(df_gold, 180)
```

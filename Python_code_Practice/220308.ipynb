```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### class 하나 생성
# 1. 인자값으로 url(주소) -> csv 파일의 주소
# 2.함수를 하나 생성-> url에 있는 csv 파일을 읽어오고, 'country 기준으로 sort, 인덱스 값을 초기화 하기
# 3.함수를 하나 생성-> 인자값을 리스트로 받는다. -> 이  매개변수를 기준으로 colmns 삭제
```

```python
class Class_1():
   
    def __init__(self, input_url):
            self.url = input_url

    def _read(self):
        self.sales = pd.read_csv(self.url)
        self.sales.sort_values("Country", inplace = True)
        self.sales.reset_index(drop = True, inplace = True) # drop 기존 인덱스 삭제,
        return self.sales
    def _drop(self, input_list=[]): # input_list는 매개변수 이다. 매개변수는 self 사용 필요 없다.
            self.sales.drop(input_list, axis = 1, inplace = True)
            return self.sales
        
    def remove_2(self, input_s_column, input_e_column):
        self.sales.drop(self.sales.loc[:, input_s_column: input_e_column], axis=1)
        return self.sales

a = Class_1(input_url = "./csv/Sales Records.csv")
a._read()
a._drop([])


     
        
```

# Sales 데이터로 판다스 활용

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
sales_df=pd.read_csv("./csv/Sales Records.csv")
```

```python
sales_df
sales_df.isna().sum()
```

```python
# 지금부터 해야될것
# 1. 컬럼 삭제(Ship,Date, Order ID, Sales Channel, Total revenue )
# 2. 컬럼명 수정
# 3. Country 기준으로 sort
# 4. index 값을 초기화
sales_df_3 = sales_df.drop(['Ship Date', 'Order ID', 'Sales Channel', 'Total Revenue'], axis=1)
```

```python
# 데이터프레임의 columns의 값을 변경해주기
sales_df_3.columns =["지역", "국가", "상품타입", "배송순위", "주문날짜", "판매개수", "단가", "원가", "총비용", "총이윤"]
sales_df_3
```

```python
# 국가 기준으로 sort 하기
sales_df_3=sales_df_3.sort_values("국가")
```

```python
# index 값 초기화
# 보여주기만 할뿐 바뀌지 않는다.
sales_df_3.reset_index(drop=True, inplace=True)
```

```python
sales_df_3
```

```python
# int 형식으로 타입 바꾸기
sales_df_3["총이윤"]=sales_df_3["총이윤"].astype(int)
sales_df_3
```

```python
# round 함수 사용 반올림 하기
sales_df_3["총비용"]=sales_df_3["총비용"].round(0).astype(int)
```

```python
# 소수점 한자리 수까지 두기
sales_df_3["원가"]=sales_df_3["원가"].round(0).astype(int)
```

```python
# round 함수 내림
sales_df_3["판매개수"].round(-1)
```

```python
# 날짜 타입 바꾸기
sales_df_3["주문날짜"] = pd.to_datetime(sales_df_3["주문날짜"], format="%m/%d/%Y")
```

```python
sales_df_3.head()
```

```python
# 그룹화 하고 총이윤 값을 합하기
#  sum() 함수, mean()함수, 표준편차 std()함수
sales_df_3.groupby(["지역","국가", "상품타입"]).sum()["총이윤"]
```

```python
# 한 나라에 대해서 그룹화 된것 가져오기
sales_df_3.groupby("국가").get_group("South Korea")
```

```python

sales_df_3[sales_df_3["국가"].isin(["South Korea", "China"])].groupby("국가").sum()["총이윤"]
```

```python
pd.DataFrame(sales_df_3.groupby("국가").mean()["총이윤"].loc[["China","Togo", "South Korea"]])
```

```python
# numpy 실습
import numpy as np
s1=pd.Series([1, np.nan, 3, 4, 5])
s2=pd.Series([1,2,np.nan, 4, 5])
s3=pd.Series([1,2,3 ,np.nan,5])
df=pd.DataFrame({'s1':s1, 's2':s2, 's3':s3})
df
```

```python
df.isna()
```

```python
# 결측값 제거하기
df.dropna(axis=0)
```

```python
# 각 열별 결측값 개수
df.isna().sum()
```

```python
df.dropna(axis="rows")
# 열은 df.dropna(axis="rows")
```

```python
df['s1']= df['s1']+1
df
```

```python
# 리스트로 합치기
_list=[5,4,3,2,1]
df["s2"]=df["s2"]+_list
df

# 튜플로 합치기
_tuple =(1,2,3,4,5)
df['s3'] = df['s3'] + _tuple
df
```

```python
_list=[2,1,2]
dot_result = df @ _list
dot_result
```

```python

df_2 = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], index=('s1', 's2', 's3'))

df @ df_2
```

```python
# 넘파이로 바꾸기
df.to_numpy()
```

```python
# 스칼라 내적
df.to_numpy().dot(df_2.to_numpy())
```

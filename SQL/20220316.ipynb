```python
# 여러개 동시에 replace할때 case when ~~ then ~~ end as ~~
sql_16= """
            SELECT ENAME,
            CASE WHEN JOB ="ANALYST" Then "분석가"
                 WHEN JOB ="CLERK" Then "점원"
                 WHEN JOB ="SALESMAN" Then "판매사원"
                 WHEN JOB ="MANAGER" Then "관리자"
                 ELSE "사장"
                 END as JOB
            from emp
            
        """
df_16 = _db.executeAll(sql_16)
df_16= pd.DataFrame(df_16)
df_16
```

```python
# 1. emp table과 dept을 leftjoin 조건 DEPTNO 같은 경우 
# 2. 출력 컬럼은 ENAME, JOB, SAL, DEPTNO, LOC
# 3. ENAME, LOC 그대로 출력
# 4. JOB은 각각 한글명으로 변경
# 5. SAL은 2450을 기준으로 2450 이상이면 HIGH 미만이면 LOW
# DEPTNO은 DEPTNO의 값과 dept table의 DNAME 값을 더해서 표시(DEPTNO-DNMAE 형식)
# 모든 컬럼의 이름은 한글로 변경

# 여러개 동시에 replace할때 case when ~~ then ~~ end as ~~
sql_17= """
      
SELECT emp.ENAME as "성명",
	        CASE 
	         WHEN emp.JOB ="CLERK" Then "점원"
                 WHEN emp.JOB ="SALESMAN" Then "판매사원"
                 WHEN emp.JOB ="MANAGER" Then "관리자"
                 ELSE "사장"
		End as `직종`,   if (emp.SAL>= 2450, 'HIGH', 'LOW') as `급여`, CONCAT(emp.DEPTNO, '-', dept.DNAME) as "부서번호-부서명", dept.LOC as `위치`
           
FROM emp LEFT JOIN dept on emp.DEPTNO = dept.DEPTNO
        """
df_17 = _db.executeAll(sql_17)
pd.DataFrame(df_17)
```

```python
# IN 함수 조건을 다중으로 적어줄때 사용한다.
# 아닌 값을 찾을 경우 NOT IN 사용한다.
sql_18= """
      
     SELECT * FROM emp
     WHERE  DEPTNO IN(10,20,30)    

        """
df_18 = _db.executeAll(sql_18)
pd.DataFrame(df_18)
```

```python
# 다중 쿼리 문 작성하기
sql_19= """
      
     SELECT (
             SELECT COUNT(*) FROM emp
             ) as `cnt(emp)`,
             (SELECT COUNT(*) FROM dept
                     ) as `cnt(dept)`

        """
df_19 = _db.executeAll(sql_19)
pd.DataFrame(df_19)
```

```python
# emp table은 사원 정보를 담고 있는 테이블
# dept table은 부서의 정보를 담고 있는 테이블이다.
# 부서의 이름이 'sales'인 소속 사원들의 리스트를 출력하라
# DEPTNO가 같아야 한다. 이때 where 조건 절 안에 dept의 쿼리를 한번더 적어준다.

sql_20= """
      
     SELECT  ENAME
     from emp
     where DEPTNO = (select DEPTNO from dept where dept.DNAME ="sales")
     

        """
df_20 = _db.executeAll(sql_20)
pd.DataFrame(df_20)
```

```python
# 조인으로 한번에 만들기
sql_21= """
      
     SELECT  ENAME
     from emp
     RIGHT JOIN dept
     ON emp.DEPTNO = dept.DEPTNO
     where dept.DNAME = "sales"
     

        """
df_21 = _db.executeAll(sql_21)
pd.DataFrame(df_21)
```

```python
# 부서의 위치가 'new york`이나 'DALLAS'가 아닌 소속 사원들의 리스트를 출력하라
sql_22= """
     SELECT ENAME
     from emp
     where DEPTNO IN (select DEPTNO from dept where dept.LOC NOT IN("NEW YORK", "DALLAS"))
        """
df_22 = _db.executeAll(sql_22)
pd.DataFrame(df_22)
```

```python
# between 구문 2000과 5000사이 둘다 >= <= 이다.
sql_23= """
     SELECT * FROM emp
     WHERE SAL BETWEEN 2000 AND 5000
     
        """
df_23 = _db.executeAll(sql_23)
pd.DataFrame(df_23)
```
